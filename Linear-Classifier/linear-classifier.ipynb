```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import struct
import array
from pathlib import Path
```

```python
def unpack(path_train, path_test):
    p = Path(path_train)
    q = Path(path_test)
    #Read CSV file for training and testing the data.
    if p.exists() and q.exists():
        df_train = pd.read_csv(path_train)
        df_test = pd.read_csv(path_test)
        return df_train, df_test
    else:
        print('Datasets not found')
        return None, None
```

```python
train_df, test_df = unpack('dataset/mnist_train.csv', 'dataset/mnist_test.csv')
```
